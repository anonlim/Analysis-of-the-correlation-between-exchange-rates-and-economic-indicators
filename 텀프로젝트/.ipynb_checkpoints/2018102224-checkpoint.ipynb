# (1) 주제 선정 이유

# (2) 가설 정의 
상품의 가격이 수요와 공급에 의해서 변동하는 것처럼 외환의 가격인 환율도 외환시장에서 수요와 공급에 의해서 결정된다. 수출이 늘어나거나 외국인들의 한국 여행, 외국인의 투자가 늘어나면 외환의 공급이 증가하기 때문에 환율이 떨어진다. 즉, 원화가 강세를 띤다. 만약 우리나라의 금리가 다른 나라에 비해서 상승하면 금융자산의 수익률이 높아지기 때문에 외국에서 자본이 유입된다. 그 결과 외화의 공급이 많아져서 환율이 내려가게 된다. <br>
이를 이용하여 환율에 큰 변동이 생기면 정부와 중앙 은행에서 금리를 조정하여 환율을 다시 안정화 시키는 경우가 있다. 따라서 ① 경상수지 ② 내외 물가비 ③ 내외 실질금리차 ④ 경제성장률 중 환률과 가장 상관 관계가 큰 변수는 **"③ 내외 실질금리차"** 일 것이다.


# (3) 인터넷을 통한 데이터 획득 
한국의 경상수지, 물가, 금리, 경제 성장률는 한국은행 경제 통계시스템에서 가져오고, 미국의 물가, 금리는 "TRADING ECONOMICS" 에서 가져오려고 했으나, "TRADING ECONOMICS" 에서 데이터를 받아오는 과정이 사이트 내에서 다운하는 과정이 아닌 메일로 보내주는 방식이었다. 1주일을 기다렸으나 답장이 없기도 하고 "TRADING ECONOMICS" 에서 가져오려던 데이터가 "interests" 라는 정보인데, 조사 과정에서 "interests"와 기준 금리는 좀 다른 부분이 있었다. "interests" 는 미국의 금리 중 은행이 각각 정한 금리이고, 미국 기준 금리인 "federal funds rate"는 정부가 정한 금리였다. 그래서 "TRADING ECONOMICS" 사이트가 아닌 미국 연방 은행 사이트인 "FRED Economic Data"에서 "federal funds rate"를 가져왔다.

1. 환율 데이터 (2005.01.01~2018.12.31)[분기별]

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("##데이터 원본")
exchange_rate = pd.read_excel("18.1.12 환율.xlsx")
exchange_rate.head(5)


```

2. 경상수지 데이터(2005.01.01~2018.12.31)[분기별]

```python
print("##데이터 원본")
current_account_balance=pd.read_excel("18.1.10 경상수지.xlsx")
current_account_balance.head(5)
```

3. 미국과 한국의 물가 데이터(2005.01.01~2018.12.31)[월별]

```python
print("##데이터 원본")
prices=pd.read_excel("소비자물가지수.xls")
prices.head(5)

```

4. 한국의 기준 금리 데이터(2005.01.01~2018.12.31)[분기별]

```python
print("##데이터 원본")
ko_funds_rate=pd.read_excel("한국 기준금리.xlsx")
ko_funds_rate.head(5)
```

5. 미국의 기준 금리 데이터(2005.01.01~2018.12.31)[월별]

```python
print("##데이터 원본")
usa_funds_rate=pd.read_excel("FEDFUNDS.xlsx")

usa_funds_rate.T
```

6. 경제 성장률(2005.01.01~2018.12.31)[분기별]

```python
print("##데이터 원본")
rate_of_economic_growth=pd.read_excel("18.1.15 경제성장률.xlsx")
rate_of_economic_growth.head(5)
```

# (4) 분석을 위한 데이터의 가공 

```python
#데이터 가공을 위해 생소한 데이터 프레임을 익숙한 리스트로 바꾸었다.
#모든 리스트는 2005년 1분기 부터 2018년 4분기까지 각각 56개의 항목을 갖도록 하였다.

exchange=[]
current=[]
prices_kor=[]
prices_usa=[]
ko_funds=[]
usa_funds=[]
economic_growth=[]


list_ex=exchange_rate.values.tolist()
list_cur=current_account_balance.values.tolist()
list_pr=prices.values.tolist()
list_kof=ko_funds_rate.values.tolist()
list_usaf=usa_funds_rate.values.tolist()
list_economic=rate_of_economic_growth.values.tolist()

for i in range(4,60):
    
    exchange.append(list_ex[0][i])
    current.append(list_cur[0][i])
    
    ko_funds.append(list_kof[0][i])
    
    economic_growth.append(list_economic[0][i])

##미국 금리, 한국과 미국의 물가 데이터는 분기별로 나온 것이 아니라 월별 데이터 밖에 없어서 월별 데이터를 분기별 데이터로 바꾸는 과정

for i in range(0,56):
    data=0
    for j in [0,1,2]:
        data=data+list_usaf[3*i+j][1]
        
    usa_funds.append(round(data/3,2))
    
for i in range(0,56):
    k_pr=0
    u_pr=0
    for j in [0,1,2]:
        k_pr=k_pr+list_pr[0][3*i+j+4]
        u_pr=u_pr+list_pr[1][3*i+j+4]
    prices_kor.append(round(k_pr/3,2))    
    prices_usa.append(round(u_pr/3,2))
    

    
#그래프를 그리기 위해 리스트를 다시 데이터 프레임을 만든다.

labels=['날짜','환율', '경상수지','한국 물가','미국 물가','내외 물가비','한국 기준 금리','미국 기준 금리', '내외 실질 금리차','경제성장률']

df1=pd.DataFrame(exchange)
df2=pd.DataFrame(current)
df3=pd.DataFrame(prices_kor)
df4=pd.DataFrame(prices_usa)
df5=pd.DataFrame(ko_funds)
df6=pd.DataFrame(usa_funds)
df7=pd.DataFrame(economic_growth)

#구해야 하는것은 내외 물가비와 내외 실질 금리차 이므로 이를 구한다.
#이번 프로젝트에선 미국의 값에서 한국의 값을 뺐다.
sub_prices=df4.sub(df3)
sub_funds=df6.sub(df5)

#분기 index 만들기
indexx=[]
date=""
for i in range(2005,2019):
    for j in [1,2,3,4]:
        date=str(i)+'-'+str(j)
        indexx.append(date)
indexxx=pd.DataFrame(indexx)
frame=pd.concat([indexxx,df1,df2,df3,df4,sub_prices,df5,df6,sub_funds,df7],axis=1,keys=labels)

finalframe=frame.set_index("날짜")
finalframe
```

#### 환율과 (경상수지, 내외 물가비, 내외 실질 금리차, 경제 성장률)을 상관 관계 분석한다. 

```python
del finalframe['한국 물가']
del finalframe['미국 물가']
del finalframe['한국 기준 금리']
del finalframe['미국 기준 금리']

corr=finalframe.corr(method='pearson')
corr
```

```python
corr[['환율']]
```

```python
print("#환율과 경상수지 의 산포도 그래프")
plt.scatter(df1,df2)
plt.show()
print("상관 계수 =",corr.iloc[1,0])
print()
print("#환율과 내외 물가비 의 산포도 그래프")
plt.scatter(df1,sub_prices)
plt.show()
print("상관 계수 =",corr.iloc[2,0])
print()
print("#환율과 내외 실질 금리차 의 산포도 그래프")
plt.scatter(df1,sub_funds)
plt.show()
print("상관 계수 =",corr.iloc[3,0])
print()
print("#환율과 경제 성장률 의 산포도 그래프")
plt.scatter(df1,df7)
plt.show()
print("상관 계수 =",corr.iloc[4,0])
print()
```

### 조사 결과에 따르면,<br><br>1. 상관 계수 값에 따라 환율과 내외 실질 금리차 가 뚜렷한 음적 상관 관계를 보임을 알 수 있고, 4개의 항목 중에 가장 상관 관계가 크다는 것을 알 수 있다.<br> <br>따라서 가설은 참으로 밝혀졌다.<br><br>2. 내외 실질 금리차가 가장 환율과 관계가 있고, 그 다음은 뚜렷한 양적 상관 관계를 보이는 경상수지 였다.<br><br>3. 내외 물가비와 경제 성장률은 위 두 원인에 비하면, 비교적 약한 상관 관계를 보임을 알 수 있다.
